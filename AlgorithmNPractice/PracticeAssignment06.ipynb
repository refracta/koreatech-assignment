### [문제 1, 179P-19] 알고리즘 4.11의 quick_select() 알고리즘을 반복구조로 기술하라.

```python
# Hoare 파티션 방법
def partition_hoare(A, left, right):
    low = left + 1
    # 좌쪽부터 증가하는 색인
    high = right
    # 우측부터 감소하는 색인
    pivot = A[left]
    # 가장 왼쪽 요소를 피봇으로 설정
    while low <= high:
        while low <= right and A[low] <= pivot: low += 1
        while high >= left and A[high] > pivot: high -= 1
        # 피봇보다 작고(low) 크지(high) 않게 low와 high 색인 변경
        if low < high:
            # low high 색인의 값 교환
            A[low], A[high] = A[high], A[low]

    A[left], A[high] = A[high], A[left]
    # high와 피벗 항목 교환하여 가장 왼쪽에 있던 피봇을 가운데에 위치시킴
    return high


# 반복 구조로 기술한 quick_select()
def quick_select_repetitive(A, left, right, k):
    pos = partition(A, left, right)
    # 분할 및 분할 색인 가져오기
    while pos + 1 != left + k:
        # 분할 색인이 정확히 k번째 수가 아닌 경우 계속 반복
        if pos + 1 > left + k:
            # 분할 색인 기준 좌측 분할에 k번째 수가 있으면 순환 호출로 left ~ pos -1의 k번째 수 구하기
            right = pos - 1
            pos = partition(A, left, right)
        else:
            # 분할 색인 기준 우측 분할에 k번째 수가 있으면 순환 호출로 pos + 1 ~ right의 k - (pos + 1 - left)번째 수 구하기
            k -= (pos + 1 - left)
            left = pos + 1
            pos = partition(A, left, right)
    return A[pos]


# 재귀 구조로 기술한 quick_select
def quick_select_recursive(A, left, right, k):
    pos = partition(A, left, right)
    # 분할 및 분할 색인 가져오기
    if pos + 1 == left + k:
        # 분할 색인이 정확히 k번째 수이면
        return A[pos]
        # 값 반환
    elif pos + 1 > left + k:
        return quick_select_recursive(A, left, pos - 1, k)
        # 분할 색인 기준 좌측 분할에 k번째 수가 있으면 순환 호출로 left ~ pos -1의 k번째 수 구하기
    else:
        return quick_select_recursive(A, pos + 1, right, k - (pos + 1 - left))
        # 분할 색인 기준 우측 분할에 k번째 수가 있으면 순환 호출로 pos + 1 ~ right의 k - (pos + 1 - left)번째 수 구하기
```

### [문제 1, 180P-21] 피벗을 이용해 리스트를 나누는 다른 알고리즘들도 있다. 알고리즘 4.12는 호어(Hoare) 분할인데, 로무토(Lomuto)분할이라는 방법도 있다. 이 방법을 조사하고 두 방법의 장단점을 비교하라. Lomuto 분할을 구현하고, 이를 이용해 알고리즘 4.11을 테스트해 보라.

```python
def partition_lomuto(A, left, right):
    pivot = A[right]
    # 최우측 요소를 피봇으로 선택
    i = left
    # 좌측 스왑용 색인를 left로 초기화

    for j in range(left, right + 1):
        # left ~ right (index 기준) 반복
        if A[j] < pivot:
            # 피봇 값보다 조회 값이 작으면
            A[j], A[i] = A[i], A[j]
            i += 1
            # 좌측 스왑용 색인의 값과 현재 조회 색인의 값을 교환 후 좌측 스왑용 색인 증가
    A[i], A[right] = A[right], A[i]
    # 피봇보다 한번 작은 수의 색인 바로 우측과 최우측에 위치한 피봇을 위치 교환
    return i
# Lomuto 분할

ARRAY = [12, 3, 5, 7, 4, 19, 26, 23, 15]
N = len(ARRAY)
# 테스트 배열 초기화

partition = partition_hoare
quick_select = quick_select_recursive
print("Hoare partition + Recursive quick select: ")
print("\t[축소정복] 3번째 작은 수: ", quick_select(ARRAY, 0, N - 1, 3))
print("\t[축소정복] 6번째 작은 수: ", quick_select(ARRAY, 0, N - 1, 6))
# Hoare 분할 + 재귀 구조

partition = partition_lomuto
quick_select = quick_select_repetitive
print("Lomuto Partition + Repetitive quick select: ")
print("\t[축소정복] 3번째 작은 수: ", quick_select(ARRAY, 0, N - 1, 3))
print("\t[축소정복] 6번째 작은 수: ", quick_select(ARRAY, 0, N - 1, 6))
# Lomuto 분할 + 반복 구조
```

Lomuto 분할은 더 간결하고 이해하기 쉽게(짧은 코드로 구현 가능) 구현할 수 있지만 Hoare 분할보다 비효율적이다.
평균적으로 Lomuto 분할은 3배 더 많은 스왑 연산을 사용하고, 배열에 같은 값의 인자가 들어오는 경우에도 Hoare 분할은 left와 right가 교차하는 순간 종료되지만 Lomuto 분할은 리스트의 길이만큼 처음부터 끝까지 반복이 진행된다.

### [LeetCode 4, 두 정렬된 배열의 중앙값]
Given two sorted arrays nums1 and nums2 of size m and n respectively, return the median of the two sorted arrays.
The overall run time complexity should be O(log (m+n)).

```python
class Solution:
    def findMedianSortedArrays(self, nums1, nums2):
        len1 = len(nums1)
        len2 = len(nums2)
        totalLen = len1 + len2
        targetLen = totalLen // 2
        # 연산에 쓰이는 길이 상수 초기화
        nums1Cursor = nums2Cursor = totalCursor = 0
        # 커서 변수 초기화
        total = [None for i in range(totalLen)]
        # nums1 + nums2 길이의 빈 배열 생성

        while nums1Cursor < len1 and nums2Cursor < len2 and totalCursor <= targetLen:
            # 둘 중 하나의 배열 순회가 모두 끝날 때 까지 반복
            # 중간 값 색인인 targetLen 까지만 값을 구하면 더 이상 반복할 필요가 없음
            # nums1[nums1Cursor]와 nums2[nums2Cursor]를 비교해서 더 작은 값을 total에 넣기
            # 선택된 배열의 커서 인덱스가 1칸 증가하게 됨
            if nums1[nums1Cursor] < nums2[nums2Cursor]:
                total[totalCursor] = nums1[nums1Cursor]
                nums1Cursor += 1
            else:
                total[totalCursor] = nums2[nums2Cursor]
                nums2Cursor += 1
            totalCursor += 1

        # 먼저 순회가 끝난 배열을 제외한 배열의 중간 지점에서 남은 값들을 순회해서 뒤로 채움
        if len1 == nums1Cursor:
            for i in range(totalCursor, targetLen + 1):
                total[i] = nums2[nums2Cursor]
                nums2Cursor += 1
        else:
            for i in range(totalCursor, targetLen + 1):
                total[i] = nums1[nums1Cursor]
                nums1Cursor += 1

        if totalLen % 2 == 0:
            # 길이가 짝수인 경우의 중간 값은 두 중간 값의 평균임
            return (total[targetLen - 1] + total[targetLen]) / 2
        else:
            # 일반적인 경우 중간 값 반환
            return total[targetLen]

```
