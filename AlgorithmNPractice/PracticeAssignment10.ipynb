## [문제 1] 호스풀 알고리즘 구현 및 테스트

- 교재 248~249쪽의 호스풀 알고리즘을 구현하라.

```python
def get_shift_table(pattern):
    m = len(pattern)
    # 패턴의 길이
    table = {}
    # 연관 배열을 이용한 시프트 테이블, lookup 과정에서 검사가 필요하지만, 아스키 코드를 포함한 모든 문자를 지원함
    for i in range(m - 1):
        table[pattern[i]] = m - 1 - i
    # 패턴의 모든 문자에 대해 해당 문자가 패턴의 몇번째 문자인지 저장
    return table

def search_horspool(T, P):
    m = len(P)
    # 패턴의 길이
    n = len(T)
    # 입력의 길이
    t = get_shift_table(P)
    # 시프트 테이블 만들기
    i = m - 1
    # 패턴의 우측 끝 문자 위치

    while i <= n - 1:
        # 모든 위치 순회
        k = 0
        # 매칭된 개수
        while k <= m - 1 and P[m - 1 - k] == T[i - k]:
            k += 1
            # 맞는 경우 k를 늘려가며 앞쪽으로 계속 진행
        if k == m:
            return i - m + 1
            # 검색 성공시 위치 반환
        else:
            i += t[T[i]] if T[i] in t else m
            # shift 값만큼 건너 뛰기, 연관 배열을 이용하므로 T[i]가 t에 없는 경우 m만큼 건너뜀
    return -1
    # 매칭 실패 -1 반환
```

```python
print("패턴의 위치 :", search_horspool("사과망고BANANAGRAPE", "BANANA"))
# 연관 배열로 시프트 테이블을 구성하였기 때문에 한글과 같은 문자에 대해서도 적용 가능
```

- 연습문제 15의 텍스트와 패턴에 대해 알고리즘을 실행하고 결과를 구하라.

```python
print("패턴의 위치 :", search_horspool("TTATAGATCTCGTATTCTTTTATAGATCTCCTATTCTT", "TCCTATTCTT"))
```

```python
# 좋은 접미어 이동 표(배열)을 만드는 함수
def get_good_suffix_shift_table(pattern):
    m = len(pattern)
    gs_shift = [0] * (m + 1);
    # 좋은 접미어 이동 배열 선언
    border_position = [0] * (m + 1);
    # 좋은 접미어 이동 배열을 구하기 위해 활용되는 경계 위치 배열

    # 전처리 1단계
    # 좋은 접미어가 패턴의 다른 곳에도 있는 경우
    i = m
    j = m + 1
    border_position[i] = j;
    while i > 0:
        while j <= m and pattern[i - 1] != pattern[j - 1]:
            if gs_shift[j] == 0:
                # 초기화되지 않은 gs_shift[j]를 패턴간 index diff 만큼으로 설정 (이동값 설정)
                gs_shift[j] = j - i
            j = border_position[j]
            # j를 j번째 경계값으로 설정
        i -= 1
        j -= 1
        border_position[i] = j
        # j를 i번째 경계값으로 설정

    # 전처리 2단계
    # 좋은 접미어의 일부만 패턴의 시작 부분에 있는 경우
    j = border_position[0]
    for i in range(m + 1):
        # 전처리 1단계의 shift 값을 덮어쓰지 않기 위해서
        if gs_shift[i] == 0:
            gs_shift[i] = j

        # j 색인은 전처리 2단계의 shift 계산간 뒤쪽의 경계값으로 유지되어야 하므로
        if i == j:
            j = border_position[j]

    # 완성된 좋은 접미어 테이블을 반환
    return gs_shift

    # https://heiwais25.github.io/algorithm/2018/10/03/boyer-moore-string-search-algorithm/
    # 위 링크의 좋은 접미어 테이블을 구하는 방법을 참고하여 알고리즘을 구현하였습니다.
```

```python
print(get_good_suffix_shift_table("HANAMANA"))
print(get_good_suffix_shift_table("NAMANA"))
# 교재에 있는 좋은 접미어 이동량 패턴 예시에 대한 이동량 테스트 계산
```

```python
def search_boyer_moore(T, P):
    m = len(P)
    # 패턴의 길이
    n = len(T)
    # 입력의 길이
    t = get_shift_table(P)
    g = get_good_suffix_shift_table(P)
    # 시프트 테이블 만들기
    i = m - 1
    # 패턴의 우측 끝 문자 위치

    while i <= n - 1:
        # 모든 위치 순회
        k = 0
        # 매칭된 개수
        while k <= m - 1 and P[m - 1 - k] == T[i - k]:
            k += 1
            # 맞는 경우 k를 늘려가며 앞쪽으로 계속 진행
        if k == m:
            return i - m + 1
            # 검색 성공시 위치 반환
        else:
            d1 = t[T[i]] if T[i] in t else m
            if k == 0:
                d = d1
            elif k > 0:
                d2 = g[-(k + 1)]
                # 좋은 접미어 배열이 색인 역순으로 배열되어 있으므로, 오른쪽 끝에서 k+1번째의 이동량 참조
                d = max(d1, d2)
            i += d
            # 교재 253쪽에 자연어로 기술된 로직, 4.5 최종 이동 거리 수식 구현
            # d = { d1          if k == 0
            #     { max(d1, d2) if k  > 0
    return -1
    # 매칭 실패 -1 반환
```
