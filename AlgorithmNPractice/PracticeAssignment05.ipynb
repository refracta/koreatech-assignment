## [문제 1] 삽입정렬을 상향식과 하향식으로 구현하고 테스트할 것
- 삽입 정렬을 상향식(알고리즘 4.3)으로 먼저 구현함.
- 연습문제 1(교재 4장 177쪽)과 같이 이 알고리즘을 하향식으로 구현(유튜브 동영상에서 코드 제공)
- 같은 데이터에 대해 테스트하고 분석할 것

```python
def printStep(arr, val):
    print(" Step %2d = " % val, end='')
    print(arr)
    # 각 스텝을 출력

def insertion_sort(A):
    n = len(A)
    for i in range(1, n):
      # 1 to 1-n까지 A 배열을 순회
        key = A[i]
        j = i - 1
        while j >= 0 and A[j] > key:
            A[j + 1] = A[j]
            j = j - 1
            # 배열 요소를 오른쪽으로 한 색인씩 이동 (Shifting)
        A[j + 1] = key
        # 순회 색인의 배열 요소가 있어야 하는 알맞는 자리에 요소를 이동
        printStep(A, i)
        # 배열 정렬 과정 출력

data = [5, 3, 8, 4, 9, 1, 6, 2, 7]
print("Original : ", data)
insertion_sort(data)
print("Insertion : ", data)

def insertion_sort_recur(A, n):
    if n == 1: return
    insertion_sort_recur(A, n - 1)
    # n == 1일때까지 재귀 호출 (상향식 방법과 결과 동일)
    key = A[n - 1]
    j = n - 2
    while j >= 0 and A[j] > key:
        A[j + 1] = A[j]
         # 배열 요소를 오른쪽으로 한 색인씩 이동 (Shifting)
        j = j - 1
    A[j + 1] = key
    # 순회 색인의 배열 요소가 있어야 하는 알맞는 자리에 요소를 이동
    printStep(A, n - 1)
    # 배열 정렬 과정 출력

data = [5, 3, 8, 4, 9, 1, 6, 2, 7]
print("Original : ", data)
insertion_sort_recur(data, len(data))
print("Insertion : ", data)
```

상향식으로 구현한 경우 삽입 정렬 알고리즘은 배열의 1부터 N-1 색인까지 순회하는 반복문 내부에

순회 색인의 배열 요소가 있어야 하는 알맞는 자리에 요소를 이동시키고, 요소가 이동되어야 하는 경우

알맞은 자리로부터 요소의 바뀌기 전 위치 직전까지의 배열 요소를 오른쪽으로 한 색인씩 이동시키는 방법을 사용하여 요소를 정렬한다.

이에 반해 하향식으로 구현한 삽입 정렬 알고리즘은 색인을 감소시킨 뒤 시행하는 재귀 반복 호출에 의해 상향식 방법의 가장 외곽 반복문과

똑같은 순서로 요소 이동 과정이 이루어지게 된다. 따라서 완전히 동일한 결과가 나오는 것을 확인 할 수 있다.


## [문제 2] 위상 정렬을 두 가지 방법으로 구현하고 테스트할 것
- 축소정복전략의 위상정렬 알고리즘 4.4를 구현할 것
- 연습문제 9(교재 4장 178쪽)과 같이 이 알고리즘을 DFS 기반의 알고리즘으로 구현할 것(유튜브 제공 않음)
- 본문의 그래프와 연습문제 7의 그래프에 대해 테스트하고 결과를 분석할 것

```python
mygraph = {
    "A": {"C", "D"},
    "B": {"D", "E"},
    "C": {"D", "F"},
    "D": {"F"},
    "E": {"F"},
    "F": {}
}
```

## 상향식 구현

```python
# 진입 차수 딕셔너리 생성
def get_in_degree_dict(graph):
    inDeg = {}
    for v in graph:
        inDeg[v] = 0

    for v in graph:
        for u in graph[v]:
            inDeg[u] += 1
    return inDeg


# 축소 정복 전략 위상 정렬 알고리즘 4.4
def topological_sort(graph):
    inDeg = get_in_degree_dict(graph)
    # 진입 차수 딕셔너리 가져오기
    vlist = [v for v in graph if inDeg[v] == 0]
    # 진입 차수가 0인 정점 필터링

    while vlist:
        v = vlist.pop()
        print(v, end=' ')
        for u in graph[v]:
            inDeg[u] -= 1
            # 방문한 정점과 연결된 정점들의 진입 차수 감소
            if inDeg[u] == 0:
                vlist.append(u)
                # 정점이 0이 된 곳 순회

print('topological_sort: ')
topological_sort(mygraph)
print()
```

## 하향식 구현

```python
def topological_sort_with_dfs(graph):
    visited = []
    inDeg = get_in_degree_dict(graph)
    # 진입 차수 딕셔너리 가져오기
    target = [v for v in graph if inDeg[v] == 0]
    # 진입 차수가 0인 정점 필터링
    print(' '.join(list(reversed(dfs_recur(graph, inDeg, visited, target)))))

def dfs_recur(graph, inDeg, visited, keys, result=[]):
    for k in keys:
        if k not in visited:
            visited.append(k)
            # 방문 정점에 추가
            for u in graph[k]:
                inDeg[u] -= 1
                # 방문한 정점과 연결된 정점들의 진입 차수 감소
                if inDeg[u] == 0:
                    dfs_recur(graph, inDeg, visited, graph[k])
                    # 정점이 0이 된 곳 순회
            result.append(k)
    return result

print('topological_sort: ')
topological_sort_with_dfs(mygraph)
print()
```

위상 정렬의 경우 축소 정복 방법과 DFS를 이용한 방법 모두 모두 방문한 정점과 연결된 정점의 진입 차수를 감소시키고

0이 된 정점을 방문하는 것을 반복하여 방문 경로를 알아내는 것은 동일하나 DFS를 이용한 방법의 경우 후위 순회를 이용하여야 이전 방문 정점에 대한 추가적인 처리 없이 효율적으로 위상 정렬을 구현할 수 있다.

하지만 후위 순회의 순회 특징 상, 마지막에 방문한 노드 순으로 처리가 이루어지기 때문에, 역순으로 출력해주는 추가적인 과정이 필요하다.

또한 두 방법의 경우 순회 순서가 다르기 때문에 서로 다른 결과가 나타날 수 있다. (한 그래프의 위상 정렬 결과에는 여러가지 경우가 있을 수 있기 때문에)
