## [문제 1] 교재 5장 연습문제 문제 3번 풀기.
- 마스터 정리를 이용해 각 복잡도 함수의 점근적 표기 구하기

### [문제 1, 227P-3] 마스터 정리를 이용해 다음 복잡도 함수의 점근적 표기를 구하라.
다음과 같은 순환식 $T(n)$이 있을 때
$
T(n)=
    \begin{cases}
        c & n=1 \\
        aT(n/b)+f(n) & otherwise \\
    \end{cases}
$

$ a \ge 1,b \ge 1 $이고, $f(n)\in O(n^d)$이라면, 다음과 같이 점근적 수행 시간을 계산할 수 있다.

$
T(n)=
    \begin{cases}
        O(n^d) & if\ a<b^d \\
        O(n^d log n) & if\ a=b^d \\
        O(n^{log _b a}) & if\ a>b^d \\
    \end{cases}
$
#### (1) $$T(n)=4T(n/2)+n, T(1)=1$$
$a=4,\ b=2$이고 $f(n)=n \in O(n)=O(n^1)$이므로 $d=1$이다. 따라서 세번째 경우인 $a>b^d$에 해당하고, 점근적 복잡도는 $O(n^{log _{2} 4}) = O(n^2)$이다.

#### (2) $$T(n)=8T(n/4)+n^2, T(1)=1$$
$a=8,\ b=4$이고 $f(n)=n^2 \in O(n^2)=O(n^2)$이므로 $d=2$이다. 따라서 첫번째 경우인 $a<b^d$에 해당하고, 점근적 복잡도는 $O(n^2)$이다.

#### (3) $$T(n)=T(n/2)+n^3, T(1)=1$$
$a=1,\ b=2$이고 $f(n)=n^3 \in O(n^3)=O(n^3)$이므로 $d=3$이다. 따라서 첫번째 경우인 $a<b^d$에 해당하고, 점근적 복잡도는 $O(n^3)$이다.

## [문제 2] 교재 5장 연습문제 9번 구현하기.
### [문제 2, 228P-9] 정렬되지 않은 리스트 A[0..n-1]이 주어졌다. 만약 이 리스트의 두 항목을 A[i], A[j]라고 할 때, 만약 i<j이고 A[i]>A[j]이면 이들은 역전(inversion)되어 있다고 정의한다. 다음 물음에 답하라.

#### (1) 리스트 A에서 역전된 항목의 수를 구하는 억지 기법 알고리즘을 설계하라. 또한 이 알고리즘의 시간 복잡도를 분석하라.

```python
def bruteforce_based_count(array):
    inversionCount = 0
    for i in range(len(array)):
        current = array[i]
        for j in range(i + 1, len(array)):
            inversionCount += 1 if current > array[j] else 0
    return inversionCount
```

중첩된 루프 내부의 `inversionCount += 1 if current > array[j] else 0`를 기본 연산이라고 하고, 배열의 길이를 n이라 하면 $\displaystyle T(n)=\sum_{i=0}^{n-1} \sum_{j=i+1}^{n-1}1$ 이고 $O(T(n))=O(n^2)$이다.

#### (2) 같은 문제에 대한 분할 정복 기법의 알고리즘을 설계하라. 이 알고리즘의 복잡도는 $0(n\ log_2 n)$ 이 될 것이다. 병합 정렬의 알고리즘과 밀접한 관련이 있다.

```python
def merge_based_count(array, left=None, right=None):
    if left is None or right is None:
        array = array.copy()
        left = 0
        right = len(array) - 1
    # 기본 값 호출의 경우 자동으로 매개변수 초기화

    count = 0
    if left < right:
        mid = (left + right) // 2
        count += merge_based_count(array, left, mid)
        count += merge_based_count(array, mid + 1, right)
        # 분할 카운팅 분기 처리 (병합 정렬과 유사)
        count += merge(array, left, mid, right)
        # 병합 정렬과 함께 역전 개수 집계
    return count


def merge(array, left, mid, right):
    count = 0
    leftCursor = left
    rightCursor = mid + 1
    sortedCursor = 0
    L = right - left + 1
    sorted = [None] * L
    # 커서와 병합 배열 초기화

    # 병합 정렬과 함께 개수 집계 루프
    while leftCursor <= mid and rightCursor <= right:
        if array[leftCursor] <= array[rightCursor]:
            # 양쪽 배열 커서에 선택된 두 요소 중 좌측 배열의 요소가 작을 때
            sorted[sortedCursor] = array[leftCursor]
            leftCursor += 1
            # 병합 배열에 좌측 요소 추가
        else:
            # 양쪽 배열 커서에 선택된 두 요소 중 우측 배열의 요소가 작을 때
            sorted[sortedCursor] = array[rightCursor]
            rightCursor += 1
            # 병합 배열에 좌측 우측 추가
            count += mid - leftCursor + 1
            # L=[→3←, 4, 5], R=[→1←, 2]와 같은 경우 [3, 4, 5]를 한꺼번에 세서 집계함
        sortedCursor += 1
        # 병합 배열 커서 이동

    if leftCursor == mid + 1:
        for i in range(sortedCursor, L):
            sorted[i] = array[rightCursor]
            rightCursor += 1
    else:
        for i in range(sortedCursor, L):
            sorted[i] = array[leftCursor]
            leftCursor += 1
    # 남은 배열 병합 배열에 붙이기

    for i in range(left, right + 1):
        array[i] = sorted[i - left]
    return count
    # 병합된 정렬 배열을 매개변수 배열에 복사
```

해당 알고리즘은 병합 정렬과 동시에 역전 개수를 집계하므로 병합 정렬과 동일한 $O(n\ log_2n)$의 시간 복잡도를 가진다.

```python
import random
ARRAY = [random.randint(0, 100) for _ in range(100)]
print("Array: ", ARRAY)
print("Count (by Bruteforce): ", bruteforce_based_count(ARRAY))
print("Count (by Divide and Conquer): ", merge_based_count(ARRAY))
```
