### [문제 1, 교재 267P] 영어 단어의 리스트를 사전순서(lexicographic order)로 정렬하는 코드를 기수 정렬로 구현 하라.

```python
# words = open('input.txt', 'r').read().split(" ")
# 단어를 파일에서 불러오는 경우

words = "good prejudice preconception bias happy glad vitiate subvert profane pervert misdirect deprave demoralize demoralise debauch debase corrupt spoil mar impair deflower violate outrage desecrate sacrilegious blasphemous secular rejoice joy infernal goddamned goddamn goddam deuced darned damned damn blessed blasted blamed blame inculpation incrimination mistake fault error pick cream clean snatch nobble kidnap abduct highly extremely full total entire fully number come amount integral intact constitutional underlying inherent organic save salve salvage relieve worst bad whip pip alienate alien shocking shameful scandalous disgraceful democratic voice vocalize vocalise sound shuddery shivery scary scarey chilling challenge wedding nuptials hymeneals trash scrap rubbish impressed well tall marvelous marvellous improbable miraculous thoroughly soundly help assistance assist aid fail flunk bomb neglect unpainful overcredulous usurious unconscionable steep outrageous extortionate exorbitant heartwarming temperamental erratic wandering planetary incompetent fumbling clumsy bungling sex adjusted killing scraps refuse garbage reprobate perverted perverse depraved unmannerly unmannered rude gladsome extralinguistic autocracy redound rh uncle pouch temporal chloromycetin chloramphenicol sage cachet abrasive abradant chlorination okay ok fine sorted fox baldwin analytic finitely sedative downer depressant coherency coherence pit lamppost thymine t goodness typo misprint literal erratum cereal healthy lycium ill macleod mystical mystic shroud pilot original archetype skeleton unquestioned undisputed unchallenged racy lively levorotatory levorotary bw bioattack stops newmarket michigan chicago boodle stockade copulatory coital lodge pyelography shipwreck landwards landward bogota rollback credit schipperke vacate resign renounce turbulent tumultuous troubled riotous disruptive peculation misappropriation misapplication embezzlement defalcation subordinate complexion freedwoman freedman quiet dependent subdue fortnightly biweekly starchlike farinaceous amyloidal amyloid amylaceous mother neology neologism coinage slack abate nernst oceanographer motormouth fright fearfulness fear veneration reverence awe kampuchean cambodian plushy anisotropically idyll soundness firmness mantilla mantelet terror panic affright senseless reasonless mindless insensible wasted superfluous purposeless pointless otiose shitwork buzz bombinate bombilate unfold stretch extend unsurpassable stunned astounded astonished astonied amazed sympathetic antithesis badly mean average apis portcullis screw love know jazz hump fuck eff bonk bed bang aperea steerable dirigible seckel phaeophyta germanite bacteriostat sissy pantywaist pansy milquetoast milksop embassy monarch capercailzie capercaillie".split(" ")
# 언어 머싱 러닝을 위한 감정 말뭉치 세트인 SentiWordNet에서 발췌


keys = '\0ABCDEFGHIJKLNMOPQRSTUVWXYZabcdefghijklnmopqrstuvwxyz';
# 기수 정렬을 위한 키 세트
def radix_sort(word_list):
    for i in reversed(range(len(max(word_list)))):
        # 단어 목록 중 가장 큰 길이를 가지는 단어의 길이만큼 순회 (max_length-1부터 0까지 순회)
        keys_dict = dict(map(lambda x: (x, []), keys))
        # {"\0": [], "A": [], "B": [], ...}의 연관 배열 생성
        for w in word_list:
            keys_dict[w[i] if len(w) > i else "\0"].append(w)
            # 순회 색인 i에 접근할 수 없는 짧은 단어이면 "\0"을, 아니면 해당 색인의 글자를 가져온 뒤
            # 가져왔던 글자 색인에 있는 연관 배열의 요소(리스트)에 단어 추가
        word_list = [inner for nested in keys_dict.values() for inner in nested]
        # {"\0": [], "A": ["Apple"], "B": ["Banana"], ...} 꼴의 연관 배열에서 색인 순서대로 값들을 얻어온 뒤
        # [["Apple"], ["Banana"], ...]꼴의 값 리스트를 순회하면서 ["Apple", "Banana"]로 평탄화
        # word_list를 갱신하고 i가 0이 될 때까지 이 작업 반복
    return word_list

print("original:", words);
print("radix_sort:", radix_sort(words));
print("sort (python):", sorted(words));
# 테스트 결과 출력

```
