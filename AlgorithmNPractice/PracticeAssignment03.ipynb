### [문제 1, 96P-20-(1)] $T(n) = 2T(n-1)\ for\ n>1,\ T(1)=4$

$T(n)=2T(n-1)$

$T(n-1)=2T(n-2)$

$T(n-2)=2T(n-3)$

$...$

$T(2)=2T(1)$

위의 양변을 곱하면

$\displaystyle\prod_{k=2}^{n} T(k) = 2^{n-1} \prod_{k=1}^{n-1} T(k)$

$\displaystyle\prod_{k=2}^{n-1} T(k)$로 양변을 나누면

$\displaystyle T(n)=T(1) \cdot 2^{n-1}=4 \cdot 2^{n-1} = 2^{n+1}$

### [문제 1, 96P-20-(5)] $T(n) = T(n/3) + n\ for\ n>1,\ T(1)=1$ ($n=3^k$라고 가정)

$T(n)=T(n/3)+n$에서 $n=3^k$으로 가정했으므로

$T(3^k)=T(3^{k-1})+3^k$ 이고

$T(3)=T(1)+3$

$T(3^{2})=T(3)+3^{2}$

$T(3^{3})=T(3^{2})+3^{3}$

$...$

$T(3^k)=T(3^{k-1})+3^k$

양변을 더하면

$\displaystyle\sum_{j=1}^{k} T(3^j) = \sum_{j=0}^{k-1} T(3^j) + \sum_{j=1}^{k} 3^j$

양변에 $ \displaystyle\sum_{j=1}^{k-1} T(3^j)$를 빼고 정리하면

$T(3^k)=T(1)+\frac{3(3^k-1)}{3-1}=1+\frac{3(3^k-1)}{3-1}=\frac{1}{2} \cdot (3^{k+1}-1)$

$3^k$를 다시 $n$으로 환원하면

$T(n)=\frac{1}{2} \cdot (3n - 1)$

### [문제 2, 96P-21] 다음과 같은 순환 알고리즘이 주어졌다. 물음에 답하라.
```python
def mystery3(n) : # n은 양의 정수
if n == 1 : return 1
else : return mystery3(n-l) + 2 * n - 1
```
#### [문제 2, 96P-21-(1)] 이 알고리즘의 반환값에 대한 순환 관계식을 만들고 풀어라. 이것은 무슨 알고리즘인가?
$mystery3(n)=T(n)$이라고 하자, `순환(재귀) 알고리즘`을 이용하여 만들어진 문제의 주어진 함수를 간단히 표현하면 다음과 같다.

$T(n) = T(n-1) + 2n -1 \ for\ n>1,\ T(1)=1$

해당 순환 관계식을 풀면

$T(2) = T(1) + 3$

$T(3) = T(2) + 5$

$T(4) = T(3) + 7$

$...$

$T(n) = T(n-1) + 2n -1$

양변을 더하면

$\displaystyle\sum_{k=2}^{n} T(k) = \sum_{k=1}^{n-1} T(k) + \sum_{k=2}^{n} (2k-1)$

$\displaystyle\sum_{k=2}^{n-1} T(k)$을 양변에 빼고 식을 정리하면

$T(n)=T(1)+n^2-1=1+n^2-1=n^2$

따라서 본 문제의 알고리즘은 $n^2$을 구하는 알고리즘이다.

#### [문제 2, 96P-21-(2)] 기본 연산으로 곱셈을 사용했을 때의 복잡도를 순환 관계식으로 만들고 풀어라. 시간 복잡도를 설명하라.
기본 연산으로 곱셈을 사용했을 때

return mystery3(n-1) + 2 `*` n - 1

n > 1일 때 모든 경우 한번의 기본 연산을 사용하게 되므로 이를 순환 관계식으로 나타내면

$T(n)=T(n-1)+1, T(1)=0$으로 표현할 수 있다.

모든 n에 따른 점화식을 나열 한 뒤 가감하여, T(n)을 구하는 방법을 이용하면

$T(n)=T(n-1)+A,\ T(1)=B$ 꼴의 점화식은 $T(n)=A(n-1)+B$로 일반화하여 나타낼 수 있다.

$A=1,\ B=0$이므로 $T(n)=n-1$이고

$O(T(n))=O(n+1)=O(n)$이므로 시간 복잡도는 $O(n)$이다.

### [문제 2, P133-5] n개의 검은 구슬과 n개의 흰 구슬이 무작위로 일렬로 나열되어 있다. 이 구슬들을 흰 구슬이 모두 먼저 나오고 이후에 검은 구슬이 나오도록 정렬하려고 한다. 이때, 구슬의 이동 은 인접한 두 구슬을 교환하는 것만 가능하다. 이 문제를 해결하기 위한 알고리즘을 작성하라.

````python
import random

marbleArray = random.choices(range(2), k=12)
marbleArray = [0, 0, 1, 1, 1, 0, 0, 1, 1, 0, 1, 0]
print(f'GENERATED ARRAY: {marbleArray}')
swapCount = 0
for i in range(len(marbleArray) - 1, 0, -1):
    for j in range(i):
        if marbleArray[j] > marbleArray[j + 1]:
            # 크기 순으로 j와 j+1 요소가 배열 되어있지 않은 경우
            printArray = marbleArray.copy()
            # 정렬 과정 출력을 위한 부분
            marbleArray[j], marbleArray[j + 1] = marbleArray[j + 1], marbleArray[j]
            # (정렬되지 않은 경우) 서로 위치 바꾸기
            swapCount += 1

            printArray[j], printArray[j + 1] = f'<{printArray[j]}>', f'<{printArray[j+1]}>'
            print(f'SWAP: {printArray} → ', end='')
            printArray[j], printArray[j + 1] = printArray[j+1], printArray[j];
            print(printArray)
            # 정렬 과정 출력을 위한 부분

print(f'RESULT ARRAY: {marbleArray}, SWAP COUNT: {swapCount}')
````

````python
import random

marbleArray = [0, 0, 1, 1, 1, 0, 0, 1, 1, 0, 1, 0]
print(f'GENERATED ARRAY: {marbleArray}')
swapCount = 0
length = len(marbleArray)
emptyCursor = length - 1
for i in range(length // 2 - 1):
  # 구슬 개수가 각각 동일할 때 작동함
  if(marbleArray[i] == 1):
      while(marbleArray[emptyCursor] != 0):
          emptyCursor -= 1;
  marbleArray[i], marbleArray[emptyCursor] = marbleArray[emptyCursor], marbleArray[i];
  swapCount += 1
  print(marbleArray) 

print(f'RESULT ARRAY: {marbleArray}, SWAP COUNT: {swapCount}')
````

````python
# 모든 경우에서 작동하는 버전
import random

marbleArray = random.choices(range(2), k=20)
print(f'GENERATED ARRAY: {marbleArray}')
swapCount = 0
length = len(marbleArray)
lCursor = 0
rCursor = length - 1
while(True):
  lrPair = marbleArray[lCursor], marbleArray[rCursor]
  if(lrPair == (1, 0)):
    marbleArray[lCursor], marbleArray[rCursor] = marbleArray[rCursor], marbleArray[lCursor]; 
    swapCount += 1
    print(marbleArray) 
    lCursor += 1
    rCursor -= 1
  elif(lrPair == (0, 1)):
    lCursor += 1
  elif(lrPair == (1, 1)):
    rCursor -= 1
  else:
    lCursor += 1

  if(lCursor > rCursor):
    break;

print(f'RESULT ARRAY: {marbleArray}, SWAP COUNT: {swapCount}')
````
